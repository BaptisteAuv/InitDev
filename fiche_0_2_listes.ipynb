## indices dans une liste

Les éléments d'une liste sont toujours numérotés dans l'ordre de la liste **à partir de la valeur 0** et de 1 en 1. On appelle ce numéro **indice**. Les indices sont donc, dans l'ordre **0, 1, 2, 3, 4, 5, 6, 7...**

```python
maliste1
```

L'élément 3 est à l'indice 0, l'élément 5.2 à l'indice 1 et l'élément 'hop' à l'indice 2.
On peut afficher la valeur d'un élément à un indice avec la syntaxe  

`nomdelaliste[indice]`

```python
maliste1[0]
```

```python
maliste1[1]
```

```python
maliste1[2]
```

```python
maliste1[3]
```
